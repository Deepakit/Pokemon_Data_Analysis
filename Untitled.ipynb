```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon=pd.read_csv("pokemon.csv")
pokemon.head()
```

```python
pokemon.info()
```

```python
pokemon.isnull().sum()
```

```python
pokemon[pokemon.duplicated('name',keep=False)].head()
```

```python
df=pokemon[pokemon['height_m'].isnull()]
print(df['generation'])
df=pokemon[pokemon['weight_kg'].isnull()]
print(df['generation'])
```

```python
pk=pokemon.groupby('generation')['height_m'].mean()


wk=pokemon.groupby('generation')['weight_kg'].mean()


#Height
pokemon.loc[pokemon['generation']==1,'height_m']=1.24
pokemon.loc[pokemon['generation']==6,'height_m']=1.09
pokemon.loc[pokemon['generation']==7,'height_m']=1.28

#Weight
pokemon.loc[pokemon['generation']==1,'weight_kg']=45.82
pokemon.loc[pokemon['generation']==6,'weight_kg']=52.225
pokemon.loc[pokemon['generation']==7,'weight_kg']=98.30

#Percentage_male
pokemon.loc[pokemon['percentage_male'].isnull(),'percentage_male']=0

#Type_2
pokemon['type2'].fillna(pokemon['type1'],inplace=True)

pokemon.isnull().sum()
```

```python
del(pokemon['base_total'])
del(pokemon['japanese_name'])
```

```python
sns.catplot(x='generation',data=pokemon,kind='count')
plt.title('Pokemons in Generations')
plt.show()
```

```python
pokemon['Against_total']=pokemon.iloc[:,1:19].sum(axis=1)
pokemon.head()
```

```python
print(pokemon['type1'].value_counts())
pokemon['type2'].value_counts()
```

```python
plt.figure(figsize=(16,10))
sns.set_style("darkgrid")
sns.countplot(x='type1',data=pokemon,hue='generation')
plt.xticks(rotation=90)
plt.title("Type-1 Pokemon")
plt.show()
plt.figure(figsize=(16,10))
sns.countplot(x='type1',data=pokemon)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(16,10))
sns.countplot(x='type2',data=pokemon)
plt.xticks(rotation=90)
plt.show()

sns.factorplot(x='generation',kind='count',data=pokemon,col='type2',col_wrap=3)
plt.show()
```

```python
pokemon.dtypes
```

```python
pokemon['strong_total']=pokemon['attack']+pokemon['defense']+pokemon['sp_attack']+pokemon['sp_defense']+pokemon['speed']
pokemon.head()
```

```python
strong_df=pokemon[['generation','Against_total','strong_total','type1','type2']]
strong_df['generation'].value_counts()
```

```python
strong_df.describe()
```

```python
strong_df.groupby('generation')[('Against_total','strong_total')].mean().sort_values(by=['Against_total','strong_total'],ascending=(True,False)).reset_index()
```

We can see Generation 7 has strongest type of pokemon Generation 3 has weakest type of Pokemon.

```python
strong_df[('type1')].value_counts()
```

```python
strong_df.groupby(['type1'])[('Against_total','strong_total')].mean().sort_values(by=['Against_total','strong_total'],ascending=(True,False)).reset_index()
```

Hence, Steel gives us strongest type of Pokemon and Rock is weakest.
